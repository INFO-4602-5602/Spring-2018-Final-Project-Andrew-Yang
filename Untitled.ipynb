```python
import numpy as np # linear algebra
import pandas as pd # data processing
import matplotlib.pyplot as plt
import seaborn as sns

# Input data files
from subprocess import check_output
# print(check_output(["ls", "../input"]).decode("utf8"))
df = pd.read_csv('database.csv');

# Fill NaN in winners with 0, for a loss
df['Winner'] = df['Winner'].fillna(0);
```

```python
dfDP = df[(df.Award == 'Cinematography')]
by_DP = dfDP.groupby('Film')
topDP = pd.DataFrame()
topDP['Wins'] = by_DP['Winner'].sum()
topDP['WinRate'] = by_DP['Winner'].mean()
topDP = topDP.sort_values('Wins', ascending=False) 
topDP['Director of Photography'] = topDP.index
#print(topact)
topact = topDP.head(20)
ax = sns.barplot(x='Wins', y='Director of Photography', color = 'gold',  
            data=topact.head(10))
plt.xlabel('Awards');
plt.ylabel('Cinematographer');
plt.title('Top 10 Oscar Winning Cinematographers');
# ax.legend_.remove()
```

```python
dfact = df[(df.Award == 'Actor') | (df.Award == 'Actor in a Supporting Role') | (df.Award == 'Actress') | (df.Award == 'Actress in a Supporting Role')]
by_act = dfact.groupby('Name')
topact = pd.DataFrame()
topact['Wins'] = by_act['Winner'].sum()
topact['WinRate'] = by_act['Winner'].mean()
topact = topact.sort_values('Wins', ascending=False) 
topact['Thespian'] = topact.index
#print(topact)
topact = topact.head(20)
ax = sns.barplot(x='Wins', y='Thespian', color = 'blue',  
            data=topact.head(20))
plt.xlabel('Awards');
plt.ylabel('');
plt.title('Top 20 Oscar Winning Actors and Actresses');
# ax.legend_.remove()
```

```python
dfDR = df[(df.Award == 'Directing')]

by_DR = dfDR.groupby('Film')

topDR = pd.DataFrame()
topDR['Wins'] = by_DR['Winner'].sum()
topDR['WinRate'] = by_DR['Winner'].mean()
topDR = topDR.sort_values('Wins', ascending=False) 
topDR['Director'] = topDR.index
#print(topact)
topact = topDR.head(16)
ax = sns.barplot(x='Wins', y='Director', color = 'teal',  
            data=topact.head(16))
plt.xlabel('Awards');
plt.ylabel('Director');
plt.title('16 Directors with more than one Oscar');
```
